```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
```

```python
html = requests.get('https://legislacao.fazenda.sp.gov.br/Paginas/art052.aspx').text
```

```python
soup = BeautifulSoup(html, "lxml")
print(soup.prettify())
```

```python
ps = soup.find_all('p')
ps
```

```python
artigos = []
for p in ps:
    classe = p.get('class')
    if classe == None:
        artigos.append(p.get_text())
```

```python
ncm4d = []
p = re.compile('[^d][^e][^.-][0-9][0-9][0-9][0-9][^.]')
for artigo in artigos:
    ncm4d.append( p.findall(artigo))
ncm4d
```

```python
y = 0
for item in ncm4d:
    z = 0
    for x in item:
        ncm4d[y][z] = x[3:7]
        z = z + 1
    y = y + 1
ncm4d
```

```python
ncm5d = []
p = re.compile('[0-9][0-9][0-9][0-9]\.[0-9][^0-9]')
for artigo in artigos:
    ncm5d.append(p.findall(artigo))
print(ncm5d.index(['9404.2;']))
ncm5d[44][0] = '9404.2'
ncm5d
```

```python
ncm6d = []
p = re.compile('[0-9][0-9][0-9][0-9]\.[0-9][0-9][^.]')
for artigo in artigos:
    ncm6d.append(p.findall(artigo))
y = 0
for item in ncm6d:
    z = 0
    for x in item:
        ncm6d[y][z] = x[0:7]
        z = z + 1
    y = y + 1
ncm6d
```

```python
ncm8d = []
p = re.compile('[0-9][0-9][0-9][0-9]\.[0-9][0-9]\.[0-9]{1,4}')
for artigo in artigos:
    ncm8d.append(p.findall(artigo))
ncm8d
```

```python
df = pd.DataFrame({'artigos' : artigos, 'ncm4d':ncm4d , 'ncm5d':ncm5d , 'ncm6d':ncm6d, 'ncm8d':ncm8d})
df[35:51]
```

```python
df.ncm4d = df.ncm4d.apply(lambda x : x if x != [] else np.nan)
df.ncm5d = df.ncm5d.apply(lambda x : x if x != [] else np.nan)
df.ncm6d = df.ncm6d.apply(lambda x : x if x != [] else np.nan)
df.ncm8d = df.ncm8d.apply(lambda x : x if x != [] else np.nan)
df.head(50)
```

```python
df['ncm'] = ''
for i in range(len(df.ncm)):
    df['ncm'][i] = []
    if df.ncm8d[i] == df.ncm8d[i]:
        df['ncm'][i].extend(df.ncm8d[i])
    if df.ncm6d[i] == df.ncm6d[i]:
        df['ncm'][i].extend(df.ncm6d[i])
    if df.ncm5d[i] == df.ncm5d[i]:
        df['ncm'][i].extend(df.ncm5d[i])
    if df.ncm4d[i] == df.ncm4d[i]:
        df['ncm'][i].extend(df.ncm4d[i])
    if df['ncm'][i] == []:
        df['ncm'][i] = np.nan
df[35:51]
```
